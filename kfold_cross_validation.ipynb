### diabetes dataset

```python
from sklearn.datasets import load_diabetes
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
X
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca1=pca.fit_transform(X)
pca1
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5)
for train,test in kf.split(X,y):
    Xtrain,Xtest=X[train],X[test]
    ytrain,ytest=y[train],y[test]
```

```python
print(Xtrain.shape,ytrain.shape)
print(Xtest.shape,ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(Xtrain,ytrain)
```

```python
ypred=model.predict(Xtest)
ypred
```

```python
from sklearn import metrics
import numpy as np
print("Mean Absolute Error= ",metrics.mean_absolute_error(ypred,ytest))
print("Mean Squared Error= ",metrics.mean_squared_error(ypred,ytest))
print("Root Mean Square Error= ",np.sqrt(metrics.mean_absolute_error(ypred,ytest)))
```
